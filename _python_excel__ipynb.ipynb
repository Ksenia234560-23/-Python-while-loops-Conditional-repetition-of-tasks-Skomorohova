**Пример таблицы расходов (Данные)**
* Создадим данные в следующем формате:
* Члены семьи: Имена членов семьи.
* Статьи расходов: Например, "Еда", "Жильё", "Транспорт", "Развлечения".
* Даты: Периоды расчёта.
* Суммы расходов: Значения расходов.

**Выполним это на Python:**


```python
import pandas as pd # Импорт библиотеки pandas для работы с табличными данными (DataFrame)
import numpy as np # Импорт библиотеки numpy для работы с массивами и математическими операциями
import matplotlib.pyplot as plt # Импорт библиотеки matplotlib для создания графиков и визуализаций

# Создание примера данных
data = { # Создание столбца 'Дата' с последовательностью дат, начиная с 2024-01-01, на 10 дней с шагом в 1 день.
    'Дата': pd.date_range(start='2024-01-01', periods=10, freq='D'), # Создание столбца 'Член семьи' со случайным выбором из списка имен, размером 10.
    'Член семьи': np.random.choice(['Алексей', 'Мария', 'Дмитрий', 'Общие'], size=10),  # Создание столбца 'Статья расходов' со случайным выбором из списка категорий, размером 10.
    'Статья расходов': np.random.choice(['Еда', 'Жильё', 'Транспорт', 'Развлечения'], size=10),  # Создание столбца 'Статья расходов' со случайным выбором из списка категорий, размером 10.
    'Сумма': np.random.randint(100, 1000, size=10) # Создание DataFrame (таблицы) из словаря data
}

df = pd.DataFrame(data) # Создание DataFrame (таблицы) из словаря data

# Сохраняем таблицу для анализа
print(df) # Вывод DataFrame в консоль для просмотра созданных данных. Это полезно для проверки правильности формирования таблицы.

# Вычисление общей суммы расходов
total_expenses = df['Сумма'].sum() # Вычисление общей суммы расходов путем суммирования значений в столбце 'Сумма'.

# Расходы по членам семьи
family_expenses = df.groupby('Член семьи')['Сумма'].sum()  # Группировка данных по столбцу 'Член семьи' и вычисление суммы расходов для каждого члена семьи.

# Расходы по статьям
category_expenses = df.groupby('Статья расходов')['Сумма'].sum() # Группировка данных по столбцу 'Статья расходов' и вычисление суммы расходов для каждой статьи.

# Печать результатов
print(f"Общая сумма расходов: {total_expenses}") # Вывод общей суммы расходов в консоль.
print(f"Расходы по членам семьи:\n{family_expenses}") # Вывод расходов по членам семьи в консоль (с переносом строки перед выводом результата).
print(f"Расходы по статьям:\n{category_expenses}") # Вывод расходов по статьям в консоль (с переносом строки перед выводом результата).
```

```python
import pandas as pd  # Импорт библиотеки pandas для работы с табличными данными (DataFrame)
import numpy as np  # Импорт библиотеки numpy для работы с массивами и математическими операциями
import matplotlib.pyplot as plt  # Импорт библиотеки matplotlib для создания графиков и визуализаций

# Создание примера данных
data = {
    'Дата': pd.date_range(start='2024-01-01', periods=10, freq='D'),  # Создание столбца 'Дата' с последовательностью дат, начиная с 2024-01-01, на 10 дней с шагом в 1 день.
    'Член семьи': np.random.choice(['Алексей', 'Мария', 'Дмитрий', 'Общие'], size=10),  # Создание столбца 'Член семьи' со случайным выбором из списка имен, размером 10.
    'Статья расходов': np.random.choice(['Еда', 'Жильё', 'Транспорт', 'Развлечения'], size=10),  # Создание столбца 'Статья расходов' со случайным выбором из списка категорий, размером 10.
    'Сумма': np.random.randint(100, 1000, size=10)  # Создание столбца 'Сумма' со случайными целыми числами от 100 до 999, размером 10.
}

df = pd.DataFrame(data)  # Создание DataFrame (таблицы) из словаря data

# Сохраняем таблицу для анализа
print(df)  # Вывод DataFrame в консоль для просмотра созданных данных. Это полезно для проверки правильности формирования таблицы.

# Вычисление общей суммы расходов
total_expenses = df['Сумма'].sum()  # Вычисление общей суммы расходов путем суммирования значений в столбце 'Сумма'.

# Расходы по членам семьи
family_expenses = df.groupby('Член семьи')['Сумма'].sum()  # Группировка данных по столбцу 'Член семьи' и вычисление суммы расходов для каждого члена семьи.

# Расходы по статьям
category_expenses = df.groupby('Статья расходов')['Сумма'].sum()  # Группировка данных по столбцу 'Статья расходов' и вычисление суммы расходов для каждой статьи.

# Печать результатов
print(f"Общая сумма расходов: {total_expenses}")  # Вывод общей суммы расходов в консоль.
print(f"Расходы по членам семьи:\n{family_expenses}")  # Вывод расходов по членам семьи в консоль (с переносом строки перед выводом результата).
print(f"Расходы по статьям:\n{category_expenses}")  # Вывод расходов по статьям в консоль (с переносом строки перед выводом результата).

# Круговая диаграмма по членам семьи
plt.figure(figsize=(8, 8))  # Создание фигуры (окна) для графика, размером 8x8 дюймов.
colors = plt.cm.viridis(np.linspace(0, 1, len(family_expenses))) # Создание списка цветов из палитры viridis
plt.pie(family_expenses, labels=family_expenses.index, autopct='%1.1f%%', startangle=140, colors=colors)  # Создание круговой диаграммы:
# family_expenses: данные для диаграммы.
# labels: подписи для каждого сектора (имена членов семьи).
# autopct: формат отображения процентов на секторах.
# startangle: угол, с которого начинается первый сектор (в градусах).
# colors: Указание цветов для секторов
plt.title('Доля расходов на каждого члена семьи')  # Установка заголовка диаграммы.
plt.show()  # Отображение графика.

# Круговая диаграмма по статьям расходов
plt.figure(figsize=(8, 8))  # Создание фигуры (окна) для графика, размером 8x8 дюймов.
colors = plt.cm.plasma(np.linspace(0, 1, len(category_expenses))) # Создание списка цветов из палитры plasma
plt.pie(category_expenses, labels=category_expenses.index, autopct='%1.1f%%', startangle=140, colors=colors)  # Создание круговой диаграммы:
# category_expenses: данные для диаграммы.
# labels: подписи для каждого сектора (наименования статей расходов).
# autopct: формат отображения процентов на секторах.
# startangle: угол, с которого начинается первый сектор (в градусах).
# colors: Указание цветов для секторов
plt.title('Доля расходов по статьям')  # Установка заголовка диаграммы.
plt.show()  # Отображение графика.
```

Создана таблица данных, представляющая расходы по членам семьи и статьям.

**1. Формирование таблицы данных
На основании предыдущей структуры данных, мы строим сводную таблицу, группируя расходы по членам семьи и статьям.**

```python
# Создание сводной таблицы расходов на каждого члена семьи по статьям
pivot_table = df.pivot_table(
    values='Сумма', # Указываем столбец 'Сумма' как значения, которые будем агрегировать в сводной таблице.
    index='Член семьи', # Устанавливаем столбец 'Член семьи' как индекс (строки) сводной таблицы.
    columns='Статья расходов', # Указываем столбец 'Статья расходов' как столбцы сводной таблицы.
    aggfunc='sum', # Используем функцию 'sum' для агрегации данных, то есть суммируем расходы.
    fill_value=0 # Заполняем пропущенные значения (NaN), которые могут возникнуть в сводной таблице, нулями.
)

# Отображение сводной таблицы
print("Сводная таблица: Расходы на каждого члена семьи по статьям") # Выводим пояснительное сообщение перед таблицей.
print(pivot_table) # Выводим полученную сводную таблицу в консоль.
```

**Таблица 3: Расходы по статьям с детализацией**

В этой таблице необходимо показать подробный анализ статей расходов. Для каждой статьи расходов выполняются следующие действия:

* Группировка расходов по статьям и датам.
* Расчет суммы расходов для каждой статьи в определенные периоды.
* Визуализация данных для выявления трендов по расходам.

**Шаги реализации:**


**1. Сбор и группировка данных**

Для выполнения задач создадим сводную таблицу, которая объединит расходы по статьям и датам.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# (Предполагается, что DataFrame df уже создан и содержит данные о расходах)

# Группировка расходов по статьям и датам
detailed_expenses = df.groupby(['Дата', 'Статья расходов'])['Сумма'].sum().reset_index()
#  - `df.groupby(['Дата', 'Статья расходов'])`: Группируем DataFrame `df` по двум столбцам: 'Дата' и 'Статья расходов'. Это означает, что мы хотим сгруппировать все строки с одинаковой датой и статьей расходов вместе.
#  - `['Сумма'].sum()`: Для каждой группы, полученной на предыдущем шаге, мы применяем функцию `sum()` к столбцу 'Сумма'. Это означает, что мы суммируем все значения в столбце 'Сумма' для каждой уникальной комбинации даты и статьи расходов.
#  - `.reset_index()`: Преобразует сгруппированные результаты обратно в DataFrame, где сгруппированные столбцы ('Дата' и 'Статья расходов') становятся обычными столбцами, а не индексом.

# Переход к сводной таблице для удобства анализа
pivot_detailed = detailed_expenses.pivot_table(
    values='Сумма',  # Указываем столбец 'Сумма' в качестве значений для сводной таблицы.
    index='Дата',  # Указываем столбец 'Дата' в качестве индекса (строк) сводной таблицы.
    columns='Статья расходов',  # Указываем столбец 'Статья расходов' в качестве столбцов сводной таблицы.
    aggfunc='sum',  # Используем функцию 'sum' для агрегации значений. Если для определенной даты и статьи расходов есть несколько записей, они будут суммированы.
    fill_value=0  # Заполняем отсутствующие значения (NaN), которые могут возникнуть в сводной таблице, значением 0. Это делается для удобства дальнейшего анализа и визуализации.
)

# Отображение таблицы
print("Таблица 3: Расходы по статьям с детализацией")  # Выводим заголовок таблицы.
print(pivot_detailed)  # Выводим полученную сводную таблицу.  Она показывает сумму расходов по каждой статье для каждой даты.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# (Предполагается, что DataFrame df уже создан и содержит данные о расходах)

# Группировка расходов по статьям и датам
detailed_expenses = df.groupby(['Дата', 'Статья расходов'])['Сумма'].sum().reset_index()
# `df.groupby(['Дата', 'Статья расходов'])`: Группируем DataFrame `df` по двум столбцам: 'Дата' и 'Статья расходов'. Это означает, что мы хотим сгруппировать все строки с одинаковой датой и статьей расходов вместе.
# `['Сумма'].sum()`: Для каждой группы, полученной на предыдущем шаге, мы применяем функцию `sum()` к столбцу 'Сумма'. Это означает, что мы суммируем все значения в столбце 'Сумма' для каждой уникальной комбинации даты и статьи расходов.
# `.reset_index()`: Преобразует сгруппированные результаты обратно в DataFrame, где сгруппированные столбцы ('Дата' и 'Статья расходов') становятся обычными столбцами, а не индексом.

# Переход к сводной таблице для удобства анализа
pivot_detailed = detailed_expenses.pivot_table(
    values='Сумма',  # Указываем столбец 'Сумма' в качестве значений для сводной таблицы.
    index='Дата',  # Указываем столбец 'Дата' в качестве индекса (строк) сводной таблицы.
    columns='Статья расходов',  # Указываем столбец 'Статья расходов' в качестве столбцов сводной таблицы.
    aggfunc='sum',  # Используем функцию 'sum' для агрегации значений. Если для определенной даты и статьи расходов есть несколько записей, они будут суммированы.
    fill_value=0  # Заполняем отсутствующие значения (NaN), которые могут возникнуть в сводной таблице, значением 0. Это делается для удобства дальнейшего анализа и визуализации.
)

# Отображение таблицы
print("Таблица 3: Расходы по статьям с детализацией")  # Выводим заголовок таблицы.
print(pivot_detailed)  # Выводим полученную сводную таблицу.  Она показывает сумму расходов по каждой статье для каждой даты.

# Построение линейного графика
colors = plt.cm.tab10(np.arange(pivot_detailed.shape[1]))  # Задаём цветовую палитру (tab10) для линий
pivot_detailed.plot(figsize=(12, 7), color=colors)  # Строим линейный график:
# figsize=(12, 7): Указываем размер графика (ширина и высота в дюймах).
# color=colors: Передаём список цветов, чтобы каждая линия имела свой цвет.
plt.title('Динамика расходов по статьям')  # Устанавливаем заголовок графика.
plt.xlabel('Дата')  # Устанавливаем подпись оси X (горизонтальной оси).
plt.ylabel('Сумма расходов')  # Устанавливаем подпись оси Y (вертикальной оси).
plt.legend(title='Статьи расходов')  # Добавляем легенду с заголовком "Статьи расходов".
plt.grid(True)  # Включаем отображение сетки для лучшей читаемости.
plt.tight_layout()  # Автоматически корректируем расположение элементов графика, чтобы они не перекрывались.
plt.show()  # Отображаем график.
```

**Для анализа доходов необходимо:**

* Создать таблицу с информацией о доходах каждого члена семьи по периодам.
* Провести агрегацию данных: суммировать доходы за указанные периоды.
* Выполнить визуализацию динамики доходов, чтобы выявить изменения и тренды.

**Шаги реализации:**

1. Подготовка данных о доходах

Создадим таблицу, аналогичную таблице расходов, с добавлением временного периода.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Пример данных о доходах семьи
income_data = {
    'Дата': pd.date_range(start='2024-01-01', periods=12, freq='M'),
    # Создаем столбец 'Дата' с последовательностью дат, начиная с 2024-01-01, на 12 месяцев с шагом в 1 месяц.
    'Член семьи': np.random.choice(['Алексей', 'Мария', 'Дмитрий', 'Общие'], size=12),
    # Создаем столбец 'Член семьи', случайным образом выбирая одного из членов семьи (Алексей, Мария, Дмитрий, Общие) для каждого месяца. Размер выборки - 12.
    'Доход': np.random.randint(500, 5000, size=12)
    # Создаем столбец 'Доход' со случайными целыми числами от 500 до 5000. Размер выборки - 12.
}

df_income = pd.DataFrame(income_data)
# Создаем DataFrame `df_income` из словаря `income_data`.  DataFrame - это табличная структура данных в pandas, похожая на таблицу в электронной таблице.

# Отображение данных
print("Данные о доходах семьи:")
# Выводим текстовое сообщение в консоль.
print(df_income)
# Выводим содержимое DataFrame `df_income` в консоль, чтобы можно было просмотреть созданные данные о доходах.
```

**2. Группировка данных по временным периодам**

Сгруппируем данные по членам семьи и месяцам для анализа доходов.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# (Предполагается, что DataFrame df_income уже создан и содержит данные о доходах)

# Группировка доходов по членам семьи и периодам
income_by_family = df_income.groupby(['Дата', 'Член семьи'])['Доход'].sum().unstack(fill_value=0)
# `df_income.groupby(['Дата', 'Член семьи'])`: Группируем DataFrame `df_income` по столбцам 'Дата' и 'Член семьи'.  Это означает, что мы хотим сгруппировать строки, имеющие одинаковую дату и члена семьи.
# `['Доход'].sum()`: Для каждой группы, полученной на предыдущем шаге, вычисляем сумму значений в столбце 'Доход'. Это означает, что мы суммируем доходы для каждого члена семьи в каждый период времени (месяц).
# `.unstack(fill_value=0)`: Преобразует сгруппированные данные в сводную таблицу, где:
# Индексом (строками) является столбец 'Дата'.
# Столбцами (заголовками столбцов) являются уникальные значения из столбца 'Член семьи'.
# Значениями в ячейках таблицы являются суммы доходов.
# `fill_value=0`: Заполняет пропущенные значения (NaN), которые могут возникнуть, значением 0.  Это необходимо, потому что не у каждого члена семьи может быть доход в каждый месяц.

# Отображение таблицы
print("Доходы семьи по периодам:")  # Выводим заголовок таблицы.
print(income_by_family)  # Выводим полученную сводную таблицу в консоль.  Таблица показывает доходы каждого члена семьи по месяцам.
```

**Таблица 5: Доходы**

**В этой таблице анализируется структура доходов семьи. Основные задачи:**
* Разделить доходы на категории (например, зарплата, дополнительный доход, инвестиции).
* Провести визуализацию данных для анализа распределения доходов

**Шаги выполнения:**
1. Подготовка данных
Создадим таблицу доходов с полями:

**Дата: Дата получения дохода**.
* Источник дохода: Категория дохода (например, "Зарплата", "Дополнительный доход").
* Член семьи: Кто получил доход.
* Сумма: Размер дохода.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Пример данных о доходах
income_data = {
    'Дата': pd.date_range(start='2024-01-01', periods=20, freq='D'),
    # Создаем столбец 'Дата', содержащий последовательность из 20 дат, начиная с 2024-01-01, с шагом в 1 день (D).
    'Источник дохода': np.random.choice(['Зарплата', 'Дополнительный доход', 'Инвестиции'], size=20),
    # Создаем столбец 'Источник дохода', содержащий случайный выбор из списка ('Зарплата', 'Дополнительный доход', 'Инвестиции') для каждой из 20 дат.
    'Член семьи': np.random.choice(['Алексей', 'Мария', 'Дмитрий'], size=20),
    # Создаем столбец 'Член семьи', содержащий случайный выбор из списка ('Алексей', 'Мария', 'Дмитрий') для каждой из 20 дат.
    'Сумма': np.random.randint(500, 5000, size=20)
    # Создаем столбец 'Сумма', содержащий случайные целые числа от 500 до 5000 (доходы).
}

df_income = pd.DataFrame(income_data)
# Создаем DataFrame `df_income` из словаря `income_data`. DataFrame - это табличная структура данных, предоставляемая библиотекой pandas.

# Отображение данных
print("Таблица доходов:")
# Выводим текстовое сообщение в консоль, указывающее, что будет отображено.
print(df_income)
# Выводим DataFrame `df_income` в консоль для просмотра.
```

**Анализ доходов**
* Сгруппируем данные для расчёта общей суммы доходов по источникам и членам семьи.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# (Предполагается, что DataFrame df_income уже создан и содержит данные о доходах)

# Суммарные доходы по источникам
total_income_by_source = df_income.groupby('Источник дохода')['Сумма'].sum()
# `df_income.groupby('Источник дохода')`:  Группируем DataFrame `df_income` по столбцу 'Источник дохода'.  Это означает, что мы объединяем все строки с одинаковым источником дохода.
# `['Сумма'].sum()`:  Для каждой группы (для каждого источника дохода) мы суммируем значения в столбце 'Сумма'. Это вычисляет общую сумму дохода для каждого источника.
# Результат сохраняется в Series с именем `total_income_by_source`, где индексом является 'Источник дохода', а значением - общая сумма дохода из этого источника.

# Суммарные доходы по членам семьи
total_income_by_family = df_income.groupby('Член семьи')['Сумма'].sum()
# `df_income.groupby('Член семьи')`: Группируем DataFrame `df_income` по столбцу 'Член семьи'.
# `['Сумма'].sum()`: Для каждой группы (для каждого члена семьи) суммируем значения в столбце 'Сумма'. Это вычисляет общую сумму дохода для каждого члена семьи.
# Результат сохраняется в Series с именем `total_income_by_family`, где индексом является 'Член семьи', а значением - общая сумма дохода для этого члена семьи.

# Отображение результатов
print("\nСумма доходов по источникам:")  # Выводим заголовок для информации об источниках дохода.  `\n` добавляет перевод строки для разделения вывода.
print(total_income_by_source)  # Выводим Series `total_income_by_source` в консоль, показывая общую сумму дохода по каждому источнику.

print("\nСумма доходов по членам семьи:")  # Выводим заголовок для информации о доходах по членам семьи.
print(total_income_by_family)  # Выводим Series `total_income_by_family` в консоль, показывая общую сумму дохода для каждого члена семьи.
```
