```python
import pandas as pd

data = {
    'name': ['Ksenia', 'Anna', 'Charlie', 'David', 'Angelina'],
    'score': [99, 88, 67, 56, 85]
}

df = pd.DataFrame(data)

file_path = 'data.csv'
df.to_csv(file_path, index=False)

print(f"Файл '{file_path}' успешно создан.")
# Импортирует библиотеку pandas для работы с данными в формате таблиц.
# Создает словарь с данными: имена и соответствующие им оценки.
# Список имен.
# Список оценок, соответствующих именам.
# Создает DataFrame из словаря, который представляет собой таблицу с данными.
# Указывает путь к файлу, в который будет сохранен
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_path):
    try:
        data = pd.read_csv(file_path)
        print("Данные успешно загружены.")
        return data
    except FileNotFoundError:
        print("Файл не найден. Проверьте путь к файлу.")
        return None

def process_data(data):
    mean_score = data['score'].mean()
    print(f"Средний балл: {mean_score:.2f}")
    return mean_score

def visualize_data(data):
    plt.figure(figsize=(13, 5))
    plt.bar(data['name'], data['score'], color='skyblue')
    plt.xlabel('Имя')
    plt.ylabel('Детали')
    plt.title('Производительность рабочих')
    plt.axhline(y=data['score'].mean(), color='r', linestyle='--', label='Средний балл')
    plt.legend()
    plt.show()

def main():
    file_path = 'data.csv'
    data = load_data(file_path)

    if data is not None:
        process_data(data)
        visualize_data(data)

if __name__ == "__main__":
    main()
# Устанавливает размер графика.
# Создает столбчатую диаграмму для отображения оценок по именам.
# Подписывает ось X.
# Подписывает ось Y.
# Устанавливает заголовок графика.
# Добавляет горизонтальную линию, показывающую средний балл.
```

## Drive REST API

Чтобы использовать Drive API, сначала нужно выполнить аутентификацию и создать клиент API.


```python
from google.colab import auth
auth.authenticate_user()
from googleapiclient.discovery import build
drive_service = build('drive', 'v3')
# Импортируем модуль для аутентификации в Google Colab
# Выполняем аутентификацию пользователя для получения разрешений на доступ к Google Drive
# Импортируем модуль для работы с Google API
# Создаем сервисный объект для работы с Google Drive API версии 3
```

С этим клиентом мы можем использовать любые функции из <a href="https://developers.google.com/drive/v3/reference/">документации по Google Drive API</a>. Примеры представлены ниже.


### Создание файла на Диске на основе данных из Python

Прежде всего, создайте локальный файл для загрузки.

```python
with open('/tmp/to_upload.txt', 'w') as f:
  f.write('my sample file')

print('/tmp/to_upload.txt contains:')
!cat /tmp/to_upload.txt
# Открывает файл '/tmp/to_upload.txt' в режиме записи ('w').
# Если файл не существует, он будет создан.
# Если файл уже существует, его содержимое будет перезаписано.
# Записывает строку 'my sample file' в файл.
# Выводит сообщение о содержимом файла.
# Использует команду cat для отображения содержимого файла '/tmp/to_upload.txt' в терминале.
```

Загрузите его, используя метод <a href="https://developers.google.com/drive/v3/reference/files/create"><code>files.create</code></a>. Более подробная информация о загрузке файлов доступна в <a href="https://developers.google.com/drive/v3/web/manage-uploads">документации для разработчиков</a>.

```python
from googleapiclient.http import MediaFileUpload

file_metadata = {
  'name': 'Sample file',
  'mimeType': 'text/plain'
}
media = MediaFileUpload('/tmp/to_upload.txt',
                        mimetype='text/plain',
                        resumable=True)
created = drive_service.files().create(body=file_metadata,
                                       media_body=media,
                                       fields='id').execute()
print('File ID: {}'.format(created.get('id')))
# файлы в Google Drive.
# Определяет метаданные для загружаемого файла,
# включая имя файла и его тип (mimeType).
# Имя файла, которое будет отображаться в Google Drive.
# MIME-тип файла (в данном случае текстовый файл).
# Создает объект MediaFileUpload для загрузки файла,
# указывая путь к файлу, его MIME-тип и возможность возобновляемой загрузки.
```

После выполнения кода в ячейке выше на странице <a href="https://drive.google.com/">https://drive.google.com/</a> появится новый файл под названием Sample file.

### Скачивание данных из файла на Диске в Python

Скачайте файл, который мы загрузили выше.

```python
file_id = created.get('id')

import io
from googleapiclient.http import MediaIoBaseDownload

request = drive_service.files().get_media(fileId=file_id)
downloaded = io.BytesIO()
downloader = MediaIoBaseDownload(downloaded, request)
done = False
while done is False:
  # _ is a placeholder for a progress object that we ignore.
  # (Our file is small, so we skip reporting progress.)
  _, done = downloader.next_chunk()

downloaded.seek(0)
print('Downloaded file contents are: {}'.format(downloaded.read()))
# Создает запрос для получения медиа-данных файла по его идентификатору.
# Создает объект BytesIO для хранения загружаемых данных в памяти.
# Инициализирует загрузчик для загрузки файла.
# Переменная для отслеживания состояния загрузки.
# Цикл для выполнения загрузки файла, пока она не завершится.
```

Чтобы скачать другой файл, приведите переменную <code>file&#95;id</code> в соответствие с его идентификатором, таким как: 1uBtlaggVyWshwcyP6kEI-y&#95;W3P8D26sz.
