# **Циклы Python while: Условное повторение задач**

Цикл while в Python позволяет многократно выполнять блок кода до тех пор, пока выполняется заданное условие. В отличие от циклов for, которые выполняются заданное количество раз, циклы while идеально подходят для ситуаций, когда количество итераций заранее неизвестно.

Циклы — довольно полезная конструкция в Python, поэтому умение писать и использовать их — важный навык для разработчика на Python.

**Начало работы с циклами Python while**

В программировании циклы — это операторы управления потоком, которые позволяют повторять заданный набор операций несколько раз. На практике вы встретите два основных типа циклов:

* forЦиклы в основном используются для многократного повторения известного количества раз, что часто применяется при обработке наборов данных с определённым количеством элементов.
* whileЦиклы обычно используются для повторения неизвестное количество раз, что полезно, когда количество повторений зависит от заданного условия.

В Python есть оба этих цикла, и в этом руководстве вы узнаете о whileциклах. В Python вы обычно используете while циклы, когда вам нужно повторить ряд задач неизвестное количество раз.

Циклы while в Python — это составные операторы с заголовком и блоком кода, которые выполняются до тех пор, пока заданное условие не станет ложным. Базовый синтаксис цикла while показан ниже:

while condition:
    <body>

В этом синтаксисе condition — это выражение, которое вычисляет цикл для определения его истинности. Если условие истинно, то выполняется тело цикла. В противном случае цикл завершается. Обратите внимание, что тело цикла может состоять из одного или нескольких операторов, которые должны быть правильно отформатированы.

Вот более подробное описание этого синтаксиса:

* while это ключевое слово, которое инициирует заголовок цикла.
* condition это выражение, оцениваемое на истинность, которое определяет условие выхода.
* <body> состоит из одного или нескольких операторов, выполняемых на каждой итерации.

Вот краткий пример того, как можно использовать цикл while для перебора убывающей последовательности чисел:

```python
number = 5

while number > 0:
     print(number)
     number -= 1
```

В этом примере number > 0 — это условие цикла. Если это условие возвращает ложное значение, цикл завершается. Тело цикла состоит из вызова print(), который выводит значение на экран. Затем вы уменьшаете значение number. Это изменение приведёт к другому результату, когда цикл оценит условие на следующей итерации.

Цикл выполняется до тех пор, пока условие остаётся верным. Когда условие становится ложным, цикл завершается, и выполнение программы переходит к первому оператору после тела цикла. В этом примере цикл завершается, когда number достигает значения, меньшего или равного 0.

Если условие цикла не становится ложным, то у вас потенциально бесконечный цикл. Рассмотрим следующий цикл и держите пальцы рядом с комбинацией клавиш Ctrl+C, чтобы завершить его выполнение:

```python
number = 5
while number != 0:
     print(number)
     number -= 1
```

```python
number = 5
while number != 0:
     print(number)
     number -= 2
```

В этом примере условие цикла — number != 0. Это условие работает, когда вы уменьшаете number на 1. Однако если вы уменьшите его на 2, условие может никогда не стать ложным, что приведёт к потенциально бесконечному циклу. В таких случаях обычно можно завершить цикл, нажав Ctrl+C, что в большинстве операционных систем вызывает исключение KeyboardInterrupt.

Обратите внимание, что цикл while сначала проверяет своё условие, прежде чем что-либо произойдёт. Если изначально условие ложно, то тело цикла никогда не запустится:

```python
number = 0
while number > 0:
     print(number)
     number -= 1
```

В этом примере, когда цикл обнаруживает, что number не больше 0, он немедленно завершает выполнение, не заходя в тело цикла. Таким образом, тело цикла никогда не выполняется.

Теперь, когда вы знаете базовый синтаксис циклов while в Python, пришло время рассмотреть несколько практических примеров. В следующем разделе вы увидите, как эти циклы можно применять в реальных сценариях.

**Использование расширенного while Синтаксиса цикла**

Цикл while в Python обладает некоторыми расширенными функциями, которые делают его гибким и мощным. Эти функции могут быть полезны, когда вам нужно настроить цикл для выполнения определённых задач.

До сих пор вы видели примеры, в которых на каждой итерации выполняется всё тело цикла. В Python есть два ключевых слова, которые позволяют изменить это поведение:

* break: Немедленно завершает цикл. Затем выполнение программы продолжается с первого оператора, следующего за телом цикла.
* continue: Завершает только текущую итерацию. Выполнение возвращается к заголовку цикла, и оценивается условие цикла, чтобы определить, будет ли цикл выполняться снова.

Циклы while в Python также имеют дополнительный синтаксис. В них есть предложение else , которое выполняется, когда цикл завершается естественным образом, потому что условие становится истинным. В следующих разделах вы узнаете больше о том, как работают операторы break и continue, а также о том, как эффективно использовать предложение else в циклах while .

**Оператор break: Досрочный выход из цикла**

С помощью оператора break вы можете завершить выполнение цикла while и продолжить работу программы с первым оператором сразу после тела цикла.

Вот короткий сценарий, который демонстрирует, как работает оператор break:

```python
number = 6

while number > 0:
    number -= 1
    if number == 2:
        break
    print(number)

print("Loop ended")
```

Когда number становится 2, достигается оператор break и цикл полностью завершается. Выполнение программы переходит к вызову print() в последней строке скрипта.

Цикл выводит значения в обычном режиме. Когда number достигает значения 2, запускается break, завершая цикл и выводя Loop ended на экран. Обратите внимание, что 2 и 1 не выводятся.

Важно отметить, что нет смысла размещать break-операторы вне условных конструкций. Предположим, вы помещаете break-оператор непосредственно в тело цикла, не заключая его в условную конструкцию. В этом случае цикл завершится на первой итерации, возможно, без выполнения всего тела цикла.

**Оператор continue: Пропуск задач на итерации**

Далее у вас есть оператор continue . С помощью этого оператора вы можете пропустить некоторые задачи в текущей итерации при выполнении заданного условия.

Следующий скрипт почти идентичен предыдущему, за исключением оператора continue вместо break:

```python
number = 6

while number > 0:
    number -= 1
    if number == 2:
        continue
    print(number)
```

**Предложение else: Выполнение задач при естественном завершении цикла**

В Python допускается необязательное else условие в конце while циклов. Синтаксис показан ниже:

while condition:
    <body>
else:
    <body>

Код в блоке else будет выполняться только в том случае, если цикл while завершится естественным образом, не встретив оператор break . Другими словами, он выполняется, когда условие цикла становится ложным, и только тогда.

Обратите внимание, что нет особого смысла использовать else в цикле, в котором нет break-оператора. В этом случае содержимое else-блока, расположенное после цикла без отступов, будет работать так же и будет выглядеть чище.

В каких случаях может пригодиться предложение else в цикле while? Один из распространённых случаев использования else — когда нужно выйти из цикла. Рассмотрим следующий пример:

```python
import random
import time

MAX_RETRIES = 5
attempts = 0

while attempts < MAX_RETRIES:
    attempts += 1
    print(f"Attempt {attempts}: Connecting to the server...")
    # Simulating a connection scenario
    time.sleep(0.3)
    if random.choice([False, False, False, True]):
        print("Connection successful!")
        break
    print("Connection failed. Retrying...")
else:
    print("All attempts failed. Unable to connect.")
```

Этот скрипт имитирует процесс подключения к внешнему серверу. Цикл позволяет выполнить несколько попыток подключения, определяемых MAX_RETRIES константой. Если подключение выполнено успешно, то оператор break завершает цикл.

Примечание: в приведенном выше примере вы использовали модуль random для имитации успешного или неудачного подключения. Вы будете использовать этот модуль в нескольких других примерах в этом руководстве. Чтобы узнать больше о случайных данных, ознакомьтесь с руководством «Генерация случайных данных в Python (руководство)».

Если все попытки подключения не увенчались успехом, выполняется оператор else, сообщая вам, что попытки были неудачными. Запустите скрипт несколько раз, чтобы проверить результаты.

**Выполнение задач на основе условия С while циклами**
Обычно цикл while используется для ожидания, пока ресурс станет доступным, прежде чем приступить к его использованию. Это распространено в следующих сценариях:

* Ожидание создания или заполнения файла
* Проверка того, подключен ли сервер к сети
* Опрашивать базу данных до тех пор, пока запись не будет готова
* Обеспечение того, чтобы REST API был отзывчивым перед отправкой запросов

Вот цикл, который постоянно проверяет, создан ли заданный файл:

```python
import time
from pathlib import Path

filename = Path("hello.txt")

print(f"Waiting for {filename.name} to be created...")

while not filename.exists():
    print("File not found. Retrying in 1 second...")
    time.sleep(1)

print(f"{filename} found! Proceeding with processing.")
with open(filename, mode="r") as file:
    print("File contents:")
    print(file.read())
```

Цикл в этом скрипте использует метод .exists() объекта Path. Этот метод возвращает True в случае завершения работы целевого файла. Оператор not инвертирует результат проверки, возвращая True в случае, если файл не завершается. В этом случае цикл ожидает одну секунду, чтобы выполнить ещё одну итерацию и снова проверить наличие файла.

**Использование while циклов для неизвестного количества итераций**

while Циклы также отлично подходят, когда вам нужно обработать поток данных с неизвестным количеством элементов. В этом случае вы не знаете, сколько итераций потребуется, поэтому вы можете использовать цикл while с True в качестве управляющего условия. Этот метод позволяет писать циклы на Python, которые будут выполнять неизвестное количество итераций.

Для наглядности предположим, что вам нужно считать данные с датчика температуры, который непрерывно их предоставляет. Когда температура становится равной или превышает 28 градусов по Цельсию, вам следует прекратить её отслеживать. Вот цикл для выполнения этой задачи:

```python
import random
import time

def read_temperature():
    return random.uniform(20.0, 30.0)

while True:
    temperature = read_temperature()
    print(f"Temperature: {temperature:.2f}°C")

    if temperature >= 28:
        print("Required temperature reached! Stopping monitoring.")
        break

    time.sleep(1)
```

В этом цикле вы используете True в качестве условия цикла, что приводит к непрерывному выполнению цикла. Затем вы считываете данные с датчика температуры и выводите текущее значение температуры. Если температура равна или превышает 25 градусов, вы выходите из цикла. В противном случае вы ждёте секунду и снова считываете данные с датчика.

**Удаление элементов из итерируемого объекта в цикле**

Изменение коллекции во время итерации может быть рискованным, особенно если вам нужно удалить элементы из целевой коллекции. В некоторых случаях хорошим решением может быть использование while цикла.

Например, предположим, что вам нужно обработать список значений и удалить каждое значение после обработки. В этом случае вы можете использовать while цикл, как показано ниже:

```python
colors = ["red", "blue", "yellow", "green"]

while colors:
     color = colors.pop(-1)
     print(f"Processing color: {color}")
```

Когда вы оцениваете список в контексте логического выражения, вы получаете True, если он содержит элементы, и False если он пуст. В этом примере colors остаётся истинным до тех пор, пока в нём есть элементы. Как только вы удаляете все элементы с помощью метода .pop(), colors становится ложным, и цикл завершается.

**Получение пользовательского ввода С помощью while цикла**

Получение пользовательского ввода из командной строки — распространённый вариант использования while-циклов в Python. Рассмотрим следующий цикл, который принимает ввод с помощью встроенных input()-функций. Цикл выполняется до тех пор, пока вы не введёте слово "stop":

```python
line = input("Type some text: ")

while line != "stop":
    print(line)
    line = input("Type some text: ")
```

Функция input() запрашивает у пользователя ввод текста. Затем вы присваиваете результат line переменной. Условие цикла проверяет, отличается ли содержимое line от "stop", и в этом случае выполняется тело цикла. Внутри цикла вы снова вызываете input() . Цикл повторяется до тех пор, пока пользователь не введёт слово stop.

Этот пример работает, но у него есть недостаток: он излишне повторяет вызов input(). Вы можете избежать повторения, используя оператор моржа, как в приведённом ниже фрагменте кода:

```python
while (line := input("Type some text: ")) != "stop":
     print(line)
```

В этом обновлённом цикле вы получаете ввод от пользователя в переменную line с помощью выражения присваивания. В то же время выражение возвращает ввод от пользователя, чтобы его можно было сравнить со значением-маркером "stop".

**Обход итераторов С помощью while циклов**

Использование цикла while со встроенной функцией next() может быть отличным способом точного управления процессом итерации при работе с итераторами и итерируемыми объектами.

Чтобы понять, как это работает, вы перепишете цикл for с помощью цикла while вместо него. Рассмотрим следующий код:

```python
requests = ["first request", "second request", "third request"]

print("\nWith a for loop")
for request in requests:
    print(f"Handling {request}")

print("\nWith a while loop")
it = iter(requests)
while True:
    try:
        request = next(it)
    except StopIteration:
        break

    print(f"Handling {request}")
```

Циклы Python forдовольно гибкие и мощные, и обычно следует предпочесть forover, whileесли вам нужно выполнить итерацию по заданной коллекции. Однако, перевод forцикла в whileцикл, как выше, дает вам еще большую гибкость в том, как вы обрабатываете итератор.

**Эмуляция циклов Do-While**

Цикл do-while — это оператор потока управления, который выполняет свой блок кода по крайней мере один раз, независимо от того, является ли условие цикла истинным или ложным.

Если вы работали с такими языками, как C , C++ , Java или JavaScript , то вам может быть интересно, где находится цикл do-while в Python. Плохая новость в том, что в Python его нет. Хорошая новость в том, что вы можете эмулировать его с помощью whileцикла с breakоператором.

Рассмотрим следующий пример, который обрабатывает пользовательский ввод в цикле:

```python
while True:
     number = int(input("Enter a positive number: "))
     print(number)
     if not number > 0:
         break
```

Опять же, этот цикл принимает пользовательский ввод с помощью встроенной input()функции. Затем ввод преобразуется в целое число с помощью int(). Если пользователь вводит число, которое равно 0или меньше, то breakоператор запускается, и цикл завершается.

Обратите внимание, что для эмуляции цикла do-while в Python условие, завершающее цикл, помещается в конец цикла, а его тело — это breakоператор. Также важно подчеркнуть, что в этом типе цикла тело выполняется по крайней мере один раз.

**Использование whileциклов для циклов событий**

Другим распространенным и расширенным вариантом использования whileциклов в Python является создание циклов событий , которые представляют собой бесконечные циклы, ожидающие определенных действий, известных как события . Как только событие происходит, цикл отправляет его соответствующему обработчику событий .

Вот некоторые классические примеры областей, в которых можно встретить циклы событий:

* Фреймворки графического пользовательского интерфейса (GUI)
* Разработка игр
* Асинхронные приложения
* Серверные процессы

Например, скажем, вы хотите реализовать игру по угадыванию чисел. Вы можете сделать это с помощью whileцикла:

```python
from random import randint

LOW, HIGH = 1, 10

secret_number = randint(LOW, HIGH)
clue = ""

# Game loop
while True:
    guess = input(f"Guess a number between {LOW} and {HIGH} {clue} ")
    number = int(guess)
    if number > secret_number:
        clue = f"(less than {number})"
    elif number < secret_number:
        clue = f"(greater than {number})"
    else:
        break

print(f"You guessed it! The secret number is {number}")
```

В этом примере игровой цикл запускает игровую логику. Сначала он берет предположение пользователя с input(), преобразует его в целое число и проверяет, больше оно или меньше секретного числа. elseПредложение выполняется, когда предположение пользователя равно секретному числу. В этом случае цикл прерывается, и игра показывает сообщение о выигрыше.

**Изучение бесконечных whileциклов**

Иногда вы можете написать whileцикл, который естественным образом не завершается. Цикл с таким поведением обычно называют бесконечным циклом , хотя это название не совсем точное, потому что в конце вам придется каким-то образом завершить цикл.

Вы можете написать бесконечный цикл намеренно или непреднамеренно . Преднамеренные бесконечные циклы — это мощные инструменты, обычно используемые в программах, которые должны работать непрерывно до тех пор, пока не будет выполнено внешнее условие, например, игровые циклы, серверные процессы и событийно-управляемые приложения, такие как приложения с графическим интерфейсом или асинхронный код.

Напротив, непреднамеренные бесконечные whileциклы часто являются результатом какой-то логической проблемы, которая не позволяет условию цикла стать ложным. Например, это может произойти, когда цикл:

* Не удается обновить условную переменную
* Использует условие, которое логически ошибочно

В этих случаях цикл ошибочно продолжает выполняться до тех пор, пока не будет прерван извне.

В следующих разделах вы узнаете об обоих типах бесконечных циклов и о том, как подходить к ним в вашем коде. Для начала вы начнете с непреднамеренных бесконечных циклов.

**Непреднамеренные бесконечные циклы**

Предположим, вы пишете whileцикл, который никогда не заканчивается из-за внутренней ошибки. Возвращаясь к примеру в начальном разделе этого руководства, у вас есть следующий цикл, который выполняется непрерывно:

```python
number = 5
while number != 0:
     print(number)
     number -= 2
```

Чтобы завершить этот код, нужно нажать Ctrl+C , что прерывает выполнение программы с клавиатуры. В противном случае цикл будет выполняться бесконечно, поскольку его условие никогда не станет ложным. В реальном коде обычно требуется правильное условие цикла, чтобы предотвратить непреднамеренное бесконечное выполнение.

В большинстве подобных случаев можно предотвратить потенциально бесконечный цикл, исправив само условие или внутреннюю логику цикла, чтобы гарантировать, что условие станет ложным в какой-то момент выполнения цикла.

В приведенном выше примере вы можете немного изменить условие, чтобы исправить проблему:

```python
number = 5
while number > 0:
     print(number)
     number -= 2
```

На этот раз цикл не идет вниз по 0значению. Вместо этого он завершается, когда numberимеет значение, которое равно или меньше 0.

В качестве альтернативы вы можете использовать дополнительные условные операторы в теле цикла, чтобы завершить цикл, используя break:

```python
number = 5
while number != 0:
     if number <= 0:
         break
     print(number)
     number -= 2
```

Этот цикл имеет то же самое исходное условие цикла. Однако он включает отказоустойчивое условие в теле цикла, чтобы завершить его в случае, если основное условие не выполняется.

Выяснение того, как исправить непреднамеренный бесконечный цикл, во многом будет зависеть от логики, которую вы используете в условии и теле цикла. Поэтому вам следует тщательно проанализировать каждый случай, чтобы определить правильное решение.

**Преднамеренные бесконечные циклы**

Намеренно бесконечные whileциклы довольно распространены и полезны. Однако для их правильного написания требуется обеспечить надлежащие условия выхода, избежать проблем с производительностью и предотвратить непреднамеренное бесконечное выполнение.

Например, вы можете захотеть написать код для службы, которая запускается и работает вечно, принимая запросы на обслуживание. Вечно в этом контексте означает, пока вы ее не закроете.

Типичный способ написать бесконечный цикл — использовать while Trueконструкцию. Чтобы гарантировать, что цикл завершится естественным образом, следует добавить один или несколько breakоператоров, заключенных в соответствующие условия.

Этот синтаксис хорошо работает, когда у вас есть несколько причин для завершения цикла. Часто бывает чище выйти из нескольких разных мест, чем пытаться указать все условия завершения в заголовке цикла.

Чтобы увидеть эту конструкцию на практике, рассмотрим следующий бесконечный цикл, который запрашивает у пользователя его пароль:

```python
MAX_ATTEMPTS = 3

correct_password = "secret123"
attempts = 0

while True:
    password = input("Password: ").strip()
    attempts += 1

    if password == correct_password:
        print("Login successful! Welcome!")
        break

    if attempts >= MAX_ATTEMPTS:
        print("Too many failed attempts.")
        break
    else:
        print(f"Incorrect password. {MAX_ATTEMPTS - attempts} attempts left.")
```

Этот цикл имеет два выходных условия. Первое условие проверяет, верен ли пароль. Второе условие проверяет, достиг ли пользователь максимального количества попыток ввести правильный пароль. Оба условия включают breakоператор для корректного завершения цикла.
